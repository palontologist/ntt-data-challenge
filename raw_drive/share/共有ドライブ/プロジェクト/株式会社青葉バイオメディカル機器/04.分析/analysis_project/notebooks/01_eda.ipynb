```python
from pathlib import Path
import pandas as pd

ENCODINGS = ('utf-8-sig', 'utf-8', 'cp932', 'shift_jis', 'euc_jp')

def load_csv_auto(path: Path) -> tuple[pd.DataFrame, str]:
    sep = '\t' if path.suffix.lower() == '.tsv' else ','
    for enc in ENCODINGS:
        try:
            return pd.read_csv(path, encoding=enc, sep=sep), enc
        except UnicodeDecodeError:
            continue
    raise RuntimeError('Unable to decode CSV with supported encodings')
```

```python
manual_root = Path.cwd()
candidates = [manual_root / 'data/train.csv', manual_root.parent / 'data/train.csv', Path('data/train.csv')]
csv_path = next((p for p in candidates if p.exists()), candidates[0])
df, used_encoding = load_csv_auto(csv_path)
print('encoding:', used_encoding)
print('shape:', df.shape)
df.head(10)
```
